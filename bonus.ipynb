```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sqlalchemy
from sqlalchemy import create_engine

from privateinfo import username, pswrd
```

```python
engine = create_engine(f'postgresql://{username}:{pswrd}@localhost:5432/Employees')
connection = engine.connect()
```

```python
#Create a histogram to visualize the most common salary ranges for employees.
Salaries_df = pd.read_sql("select * from salaries", connection)
Salaries_df.hist(column='salary')
plt.title("Most common salary ranges for employees")
plt.xlabel("Salary Range")
plt.ylabel("Frequency")
```

We can see the data is right-skewed distributed (not Normal distributed) having the modes are higher than the actual mean of salaries

```python
#Create a bar chart of  average salary by title
plt.xlabel("Salary Range")
plt.ylabel("Frequency").
Avsalarybytittle_df = pd.read_sql('''SELECT e.emp_no, t.title, s.salary
FROM employees e
JOIN salaries s ON (e.emp_no=s.emp_no)
JOIN titles t ON(t.title_id=e.emp_title_id)
ORDER BY title''', connection)
```

```python
titles_salary= Avsalarybytittle_df.groupby("title")["salary"].mean()
titles_salary.plot.bar()
plt.title(" Average Salary by Title")
plt.xlabel("Titles")
plt.ylabel("Avg Salary")
```
